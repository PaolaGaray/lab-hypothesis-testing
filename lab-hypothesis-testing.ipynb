```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import ttest_ind

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

__Hypothesis:__
* Null Hypothesis: There is no significant difference in the average HP between _Dragon_ type and _Grass_ type Pok√©mon.
* Alternative Hypothesis: _Dragon_ type Pok√©mon have, on average, higher HP than  _Grass_ type Pok√©mon.

```python
# Filter Dragon-type and Grass-type Pok√©mon
dragon_hp = df[df['Type 1'] == 'Dragon']['HP']
grass_hp = df[df['Type 1'] == 'Grass']['HP']
```

```python
# Perform a t-test (equal_var=False allows for unequal variance between the groups)
t_stat, p_value = ttest_ind(dragon_hp, grass_hp, equal_var=False)

# For one-tailed test, divide p-value by 2
p_value_one_tailed = p_value / 2

# Output the results
print(f"T-statistic: {t_stat:.4f}")
print(f"One-tailed p-value: {p_value_one_tailed:.4f}")
```

__Conclusion:__
- The one-tailed p-value is 0.0008, which is much smaller than the 5% significance level (0.05).
- Since ùëù < 0.05, we reject the null hypothesis.

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


__Hypothesis for Each Stat:__

For each stat (HP, Attack, Defense, Sp. Atk, Sp. Def, Speed), we will set up the following hypotheses:
- Null Hypothesis: There is no significant difference in the mean stat between Legendary and Non-Legendary Pok√©mon.
- Alternative Hypothesis: There is a significant difference in the mean stat between Legendary and Non-Legendary Pok√©mon.

```python
# Load the dataset
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv"
df = pd.read_csv(url)

# Filter the data into Legendary and Non-Legendary groups
legendary = df[df['Legendary'] == True]
non_legendary = df[df['Legendary'] == False]

# List of stats to compare
stats_to_compare = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Perform independent t-tests for each stat
results = {}
for stat in stats_to_compare:
    t_stat, p_value = ttest_ind(legendary[stat], non_legendary[stat], equal_var=False)
    results[stat] = {'T-statistic': t_stat, 'P-value': p_value}

# Output the results
for stat, result in results.items():
    print(f"{stat}: T-statistic = {result['T-statistic']:.4f}, P-value = {result['P-value']:.4f}")
    
# Apply Bonferroni correction for multiple comparisons (optional)
alpha = 0.05 / len(stats_to_compare)
print(f"\nBonferroni-corrected significance level: {alpha:.4f}")
```

__Interpretation of Results:__
The t-tests were performed for six stats (HP, Attack, Defense, Sp. Atk, Sp. Def, Speed), comparing Legendary and Non-Legendary Pok√©mon. Here are the key results:

- T-statistics and p-values for all stats are extremely low (all are 0.0000), meaning the differences are highly significant.
- The Bonferroni-corrected significance level is 0.0083 (since we divided 0.05 by 6, the number of tests).

__Conclusion:__
Since all the p-values are well below the corrected threshold (0.0083), we can conclude the following:

- There is strong evidence to reject the null hypothesis for all six stats.
- This means that there are significant differences between Legendary and Non-Legendary Pok√©mon for all six stats (HP, Attack, Defense, Special Attack, Special Defense, and Speed).
In other words, Legendary Pok√©mon tend to have significantly higher stats across the board compared to Non-Legendary Pok√©mon.

**Challenge 2**

In this challenge, we will be working with california-housing data. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
# Coordinates of school and hospital
school_coords = (-118, 37)
hospital_coords = (-122, 34)

# Function to calculate Euclidean distance
def calculate_distance(lon, lat, ref_lon, ref_lat):
    return np.sqrt((lon - ref_lon)**2 + (lat - ref_lat)**2)

# Calculate distances to the school and hospital
df['distance_to_school'] = calculate_distance(df['longitude'], df['latitude'], *school_coords)
df['distance_to_hospital'] = calculate_distance(df['longitude'], df['latitude'], *hospital_coords)

# Classify houses as "close" if distance is less than 0.50 to either school or hospital
df['close_to_school_or_hospital'] = (df['distance_to_school'] < 0.50) | (df['distance_to_hospital'] < 0.50)

# Split the dataset into houses close and far from either a school or hospital
close_houses = df[df['close_to_school_or_hospital'] == True]
far_houses = df[df['close_to_school_or_hospital'] == False]

# Perform independent t-test on house prices (median_house_value)
t_stat, p_value = ttest_ind(close_houses['median_house_value'], far_houses['median_house_value'], equal_var=False)

# Output the results
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")

# Significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Houses close to a school or hospital are significantly more expensive.")
else:
    print("Fail to reject the null hypothesis: No significant difference in house prices.")
```

Interpretation of the Results:
T-statistic: -17.1742
P-value: 0.0001
Since the p-value (0.0001) is much smaller than 0.05, we can reject the null hypothesis. This means there is strong statistical evidence to support the claim that houses close to a school or hospital are, on average, significantly more expensive compared to houses that are farther away.
